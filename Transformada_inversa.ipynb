Método de la transformada inversa

Es una técnica usada para generar variables aleatorias de una distribución específica, a partir de variables aleatorias uniformes en el intervalo [0,1].
 Este método se basa en la relación entre la función de distribución acumulativa (CDF) de una variable aleatoria y su inversa.

$f(x)=1-e^{λx)$


Igualamos la función a U, resolvemos y despejamos x de forma que :\\
$X=\frac{-1}{λ} ln(1-u)$

Pseudocodigo:
 Paso 1: Generar U∼(0,1) \\
 Paso 2: Se hace $X=\frac{-1}{λ} ln(1-u)$

 Paso 3: Salir $X$


```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt

from statistics import mean, median, variance, stdev
import math
```

```python
def inversa(lamda, n):
    Lista = []
    for _ in range(n):
        u = rnd.random()
        x = -1 / lamda * np.log(1 - u)
        Lista.append(x)
    return Lista
```

```python
G=inversa(.5,1000)
print (G)
```

```python
miu = np.mean(G)
print (miu)
```

```python
sigma = np.var(G)
print (sigma)
```

```python
plt.hist(G, bins= 200)
```

```python
plt.plot(G)
```

Para la función de densidad

```python

def densidad(x, lamda):
    if x >= 0:
        return 1 - math.exp(-lamda * x)
    else:
        return 0
```

```python

lamda = 0.5  # Parámetro lambda

# Generar datos para la gráfica
x_values = np.linspace(0, .5, 100)
y_values = [densidad(x, lamda) for x in x_values]

# Graficar la función de densidad de probabilidad
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'Distribución Exponencial')

plt.show()
```

Graficamos la función de distribución acumulada

Para resolver analíticamente la función acumulada de una distribución exponencial con parámetro λ, utilizamos la definición :
Ahora graficaremos la función de distribución acumulativa
$F(x)=\int_{-∞} ^{x} 1-e^{λx}$


```python
def acumulada(x, lamda):
    if x >= 0:
        return 1 - np.exp(-lamda * x)
    else:
        return 0
```

```python
lamda = 0.5
x_values = np.linspace(0, 5, 100)
y_values = [acumulada(x, lamda) for x in x_values]
# Graficar la función de distribución acumulada (CDF)
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=f'Distribución Exponencial, $\lambda$={lamda}')
plt.title('Función de Distribución Acumulada ')

plt.show()
```
