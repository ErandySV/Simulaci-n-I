El método de aceptación y rechazo es una técnica para generar muestras de una distribución f(x) a partir de una distribución propuesta, se utiliza cuando la función de densidad
f(x) es difícil de muestrear directamente

Como propuesta tomaremos g(x) como una distribución simple, siendo una distribución uniorma en [0,1]

La constante la tomaremos como: $C=e^-1/1$ dado \\
$\int_{0}^{1} (e^-x dx)=1-e^-1$

Pseudocodigo:\\
Paso 1: Generar $U∼(0,1)$\\
Paso 2: Se calcula $f(x)$ como $ \frac{e^-x}{e-1}$\\
Paso 3: Se acepta si $ U\leq \frac{f(x)}{cg(x)}$

```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean, median, variance, stdev
import math
```

```python
def aceptarechaza(n):

    muestra = []
    c = np.exp(1) - 1

    while len(muestra) < n:
        U = rnd.uniform(0, 1)
        X = rnd.uniform(0, 1)

        f_X = np.exp(-X) / (np.exp(1) - 1)
        g_X = 1  # Uniforme en [0,1], propuesta

        # Se acepta si
        if U <= f_X / (c * g_X):
            muestra.append(X)

    return muestra
```

```python
G=aceptarechaza(1000)
print(G)
```

```python
plt.hist(G)
```
