```python
import random as rnd                                # importamos el módulo random
import numpy as np                             # importamos numpy con un alias (np)
import matplotlib.pyplot as plt                 # ídem
#from random import expovariate, gauss
from statistics import mean, median, variance, stdev      # exportamos funciones especificas de statistics
```

```python

def expo(lamda = 2,n = 10):
   L = -1/lamda
   lista = []
   for i in range(n):
      u=rnd.random()
      X=(L*np.log(u))
      lista.append(X)
   return lista
```

```python
G=expo(.5,100000)
print (G)
```

```python
miu = np.mean(G)
print (miu)
```

```python
sigma = np.var(G)
print (sigma)
```
