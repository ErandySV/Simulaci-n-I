 Experimento de la aguja del Buffon \\


Este problema denominado "La aguja de Buffon" relaciona el número pi con el lanzamiento de una aguja sobre una superficie rayada.

Buffon demostro que si lanzamos, al azar, una aguja de longitud L sobre una superficie en la que hay dibujadas líneas paralelas separadas una distancia D, la probabilidad de que la aguja corte a una línea es : \\
$\frac{L*\pi}{D*2} =P$ \\
$\frac{L*\pi}{D*2} =\frac{n}{N}$ \\
Vamos a utilizar este resultado para medir $\pi$ \\
Si despejamos $\pi$ tenemos : \\

$\pi=P * \frac{D*2}{L} $  \\
$\pi=(\frac{n}{N})  \frac{D*2}{L} $  
En donde: \\
 $n ⟶ $ Número de veces que la aguja corta la línea \\
$N ⟶$ Número total de agujas  \\
$L ⟶$ Longitud de la aguja \\
$D ⟶$ Distancia entre lineas

```python
import random
import math
import matplotlib.pyplot as plt

def buffon(N, L, D):
    fig, ax = plt.subplots()
    for i in range(0, int(5 / D) + 2):
        ax.axhline(i * D, color='black', linestyle='-', linewidth=1)
    n = 0
    Lista =[]
    for _ in range(N):
        #  centro de la aguja
        x_centro = random.uniform(0, 5)
        y_centro = random.uniform(0, 5)
        theta = random.uniform(0, math.pi)

        #  extremos de la aguja
        x_end1 = x_centro + (L / 2) * math.cos(theta)
        y_end1 = y_centro + (L / 2) * math.sin(theta)
        x_end2 = x_centro - (L / 2) * math.cos(theta)
        y_end2 = y_centro - (L / 2) * math.sin(theta)
        if int(y_end1 // D) != int(y_end2 // D):
            color = 'red'
            n += 1
            Lista. append (n/(i+1))

        else:
            color = 'black'

        # Dibujar la aguja
        ax.plot([x_end1, x_end2], [y_end1, y_end2], color=color)

    plt.show()
    plt.plot(Lista)

    # Estimación
    estimacion = (2 * L * N) / (n * D)
    return estimacion
```
