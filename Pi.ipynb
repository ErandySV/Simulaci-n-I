Estimación del número π

El uso del método de Monte Carlo para aproximar el valor de $\pi$ consiste en dibujar un cuadrado y detro de ese cuadrado, diujar un circulo con diámetro de igual medida que el lado del cuadrado \\
Se dibujan puntos de manera aleatoria sobre toda la superficie dibujada y se divide el número de puntos dentro de la figura entre el total de puntos hechos

Para modelarlo con simulación generamos $\frac{1}{4}$ del circulo, con radio 1 \\
$y=\sqrt{1-x^2} $ con $0\leq x\leq 1$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 1000)
y = np.sqrt(1 - x**2)


plt.plot(x, y, label='$y = \sqrt{1 - x^2}$', color='b', linewidth=2)
plt.show()
```

Experimento: Generar $u_1,u_2∼U(0,1)$ \\
Exito: $u_2\leq \sqrt{1-u^2}$ \\
Repetir N veces \\
$\frac{A_0}{A}≈\frac{n}{N}$ \\
Si $A=\frac{\pi}{4}$ en este caso \\
entonces $4 *\frac{n}{N} ≈ \pi$

El código quedaría de la siguiente forma

```python

import numpy as np
import random
import matplotlib.pyplot as plt
N=650
exito=0
lista=[]
for i in range(N):
 x = np.linspace(0,1,100)
 u_1=random.uniform(0,1)
 u_2=random.uniform(0,1)

 plt.plot(x,np.sqrt (1-x**2) )
 if u_2<np.sqrt (1-u_1**2):
   plt.scatter(u_1,u_2, color="red")
   exito+=1
   lista. append (exito/(i+1))
 else:
   plt.scatter (u_1, u_2, color="blue")
plt.show( )
estimacion=4*(exito/N)
plt.plot(lista)

print("La estimación es:",estimacion)
```

Se ve que una muestra de tamaño $N ≈ 500 da un resultado aceptable
