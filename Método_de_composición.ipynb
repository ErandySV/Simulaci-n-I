Es una técnica de generación de variables aleatorias utilizada en la simulación y estadística para crear muestras de una distribución compleja a partir de la combinación de muestras de distribuciones más simples. Esta técnica se basa en descomponer una distribución compleja en varias componentes (distribuciones más simples) y generar muestras según una mezcla de estas componentes

```python
import random as rnd                                # importamos el módulo random
import numpy as np                             # importamos numpy con un alias (np)
import matplotlib.pyplot as plt                 # ídem
#from random import expovariate, gauss
from statistics import mean, median, variance, stdev
import math
from math import sqrt
```

```python
import random as rnd
from math import sqrt

def cp(n, a, b, c):
    A1 = (b-a)/(c-a)
    X = []  # Lista para almacenar los valores
    for i in range(n):
        R1 = rnd.random()
        R2 = rnd.random()

        if R1 >= 0 and R1 <= A1:
            x = a + (c - a) * sqrt(R2)
        else:
            x = c - (c - b) * sqrt(R2)

        X.append(x)
    return X
```

```python
G= cp(100, 10, 15, 7.5)
print (G)
```

```python
plt.hist(G, bins=30, edgecolor='black')
plt.show()
```
